```python
import csv as csv
```

```python
My_File = csv.reader(open('../Compare Results/CSV/genderbasemodel.csv', 'r'))
Correct_File = csv.reader(open('../Compare Results/CSV/gendermodel.csv', 'r'))

Header_Skip = My_File.__next__()
Header_Skip = Correct_File.__next__()
```

```python
for row in My_File:
    Correct_File_Row = Correct_File.__next__()
    if row != Correct_File_Row:
        print('These Lines did not match')
        print("My File Row:", My_File_Row)
        print("Correct File Row:", Correct_File_Row)
    if row == Correct_File_Row:
        print("Correct")
        print(row)
        print(Correct_File_Row)
```
